```python
import matplotlib.pyplot as plt
import statistics

x = [5,7,8,7,2,17,2,9,4,11]            # ages of a car model
y = [99,86,87,88,111,86,103,87,94,78]  # maximum speed

#get the slope and intercept.
slope, intercept = statistics.linear_regression(x, y)

print("Slope = ", slope)
print("Intercept = ", intercept)
```

```python
def myfunc(x):
  y = slope * x + intercept
  return y

#a new array with new values for the y-axis
mymodel = list(map(myfunc, x))
print(mymodel)

#plot the data points and the regression line
plt.scatter(x, y)                   #plot the data points
plt.plot(x, mymodel, color="gray")  #plot the regression line
plt.show()
```

```python
#Predict the speed of a 12 years old car
speed = myfunc(12)
print("The speed of a 12 years old car is",speed)

#plot the data points and the regression line
plt.scatter(x, y)                  #plot the data points
plt.plot(x, mymodel, color="gray") #plot the regression line

#plot the speed prediction (when x=12)
plt.plot(12,speed, marker='o', color='red')
plt.show()
```

```python
import statistics
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_data.csv')

x=df['Duration']  #x-axis series of data
y=df['Calories']  #y-axis series of data

#get the slope and intercept.
slope, intercept = statistics.linear_regression(x, y)
print("Slope = ", slope)
print("Intercept = ", intercept)

#define a linear regression model
def myfunc(x):
  y = slope * x + intercept
  return y
#a new array with new values for the y-axis
mymodel = list(map(myfunc, x))

#plot the data points and the regression line
plt.scatter(x, y)                  #plot the data points
plt.plot(x, mymodel, color="pink") #plot the regression line

#Predict the calory of 120 minutes exercises
calory = myfunc(120)
print("The calories burned of a 120 minutes exercise is",calory)
#Plot the predicted point
plt.plot(120, calory, color="red", marker="x")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('/content/topBrands2022.csv')

# Strip any spaces from the column names just in case
df.columns = df.columns.str.strip()

# Check the column names and first few rows to confirm
print(df.columns)
print(df.head())

# Now you can run the tasks, for example:

# Task 1: Frequency distribution for 'industry_type' and bar chart
industry_counts = df['industry_type'].value_counts()
print(industry_counts)

industry_counts.plot(kind='bar')
plt.title('Frequency of Industry Type')
plt.xlabel('Industry Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
industry_counts = df['industry_type'].value_counts()
print(industry_counts)

industry_counts.plot(kind='bar')
plt.title('Frequency of Industry Type')
plt.xlabel('Industry Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
bins = [0, 20, 40, 60, 80, 100, 300]
labels = ['0-20', '20-40', '40-60', '60-80', '80-100', '100-300']

df['Brand_Value_Bin'] = pd.cut(df['brand_value_in_billions_USD'], bins=bins, labels=labels)
value_counts = df['Brand_Value_Bin'].value_counts().sort_index()
print(value_counts)

value_counts.plot(kind='bar')
plt.title('Brand Value Frequency Distribution')
plt.xlabel('Brand Value Range (Billion USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
bins = [0, 20, 40, 60, 80, 100, 120, 140, 160, 300]
labels = ['0-20', '20-40', '40-60', '60-80', '80-100', '100-120', '120-140', '140-160', '160-300']

df['Brand_Revenue_Bin'] = pd.cut(df['brand_revenue_in_billions_USD'], bins=bins, labels=labels)
revenue_counts = df['Brand_Revenue_Bin'].value_counts().sort_index()
print(revenue_counts)

revenue_counts.plot(kind='bar')
plt.title('Brand Revenue Frequency Distribution')
plt.xlabel('Brand Revenue Range (Billion USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Conclusion
#Tech companies rank the highest.
#In brand value range and brand revenue range, 0-20 billion USD is the highest.
```

```python
crosstab_value = pd.crosstab(df['industry_type'], df['Brand_Value_Bin'])
print(crosstab_value)
```

```python
sns.heatmap(crosstab_value, annot=True, cmap='YlGnBu')
plt.title('Industry Type vs Brand Value')
plt.xlabel('Brand Value (Billion USD)')
plt.ylabel('Industry Type')
plt.show()
```

```python
bins_revenue = [0, 20, 40, 60, 80, 100, 400]
labels_revenue = ['0-20', '20-40', '40-60', '60-80', '80-100', '100-400']

df['Brand_Revenue_Bin2'] = pd.cut(df['brand_revenue_in_billions_USD'], bins=bins_revenue, labels=labels_revenue)
crosstab_revenue = pd.crosstab(df['industry_type'], df['Brand_Revenue_Bin2'])
print(crosstab_revenue)
```

```python
sns.heatmap(crosstab_revenue, annot=True, cmap='OrRd')
plt.title('Industry Type vs Brand Revenue')
plt.xlabel('Brand Revenue (Billion USD)')
plt.ylabel('Industry Type')
plt.show()
```

```python
#Conclusion
#Tech companies have the highest frequency in most brand revenue ranges, especially in the highest range 100-400 billion USD.
```
